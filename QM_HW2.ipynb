## Punto 4 B

La solución más general para la ecuación de Schrödinger es: \\

$\Psi(x,t) = \sum_{n=1}^{\infty}c_n \sqrt{\frac{2}{a}}sin\left(\frac{n\pi}{a}x\right) e^{-i(n^2 \pi^2 \hbar/2ma^2)t}$ \\
Recordando que: \\
$c_n = 0$ para n pares \\
$c_n = \left(\frac{2}{\pi n} \right)^3 \sqrt{15}$ para n impares

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from numpy import pi, sqrt, sin, exp
```

```python
a = 1
hbar = 1
m = 1

x_values = np.linspace(0, a, 100)
t_values = np.linspace(0, 2 * pi, 100)
x, t = np.meshgrid(x_values, t_values)


def wave_f(n, t):
  prefactor = ((2 / (pi * n)) ** 3) * sqrt(15) * sqrt(2 / a)
  spatial_term = sin((n * pi / a) * x)
  temporal_term = exp(-1j * (n ** 2 * pi ** 2 * hbar / (2 * m * a ** 2)) * t)
  wave_function = prefactor * spatial_term * temporal_term
  return wave_function

wave_f1 = wave_f(1, t_values)
wave_f2 = wave_f(2, t_values)
wave_f3 = wave_f(3, t_values)
wave_f4 = wave_f(4, t_values)

fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, t, np.real(wave_f1), cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Re[ψ(x, t)]')
ax.set_title('Wave Function ' + '$\psi_{1}$' + '(x, t)')

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(x, t, np.real(wave_f2), cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Re[ψ(x, t)]')
ax.set_title('Wave Function ' + '$\psi_{2}$' + '(x, t)')

plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, t, np.real(wave_f3), cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Re[ψ(x, t)]')
ax.set_title('Wave Function ' + '$\psi_{3}$' + '(x, t)')

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(x, t, np.real(wave_f4), cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Re[ψ(x, t)]')
ax.set_title('Wave Function ' + '$\psi_{4}$' + '(x, t)')

plt.show()
```

```python
x_values = np.linspace(0, a, 400)
x, t = np.meshgrid(x_values, t_values)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()

for idx, n in enumerate([1, 2, 3, 4]):
    prefactor = ((2 / (pi * n)) ** 3) * sqrt(15) * sqrt(2 / a)
    spatial_term = sin((n * pi / a) * x_values)
    prob_distribution = np.abs(prefactor * spatial_term) ** 2

    axs[idx].plot(x_values, prob_distribution, label=f'n = {n}')
    axs[idx].set_xlabel('Position (x)')
    axs[idx].set_ylabel('Probability Density |ψ(x, t)|^2')
    axs[idx].legend()
    axs[idx].set_title(f'Probability Distribution for n = {n}')

plt.tight_layout()
plt.show()
```
